```python
province_case = pd.read_csv("Indonesia_coronavirus_daily_data.csv")
province_case.head()
```

```python
case_province = indo.merge(province_case, left_on="PROVINSI", right_on="Province")
case_province.head()
```

<h1>FEATURE SELECTION

```python
daily_case = case_province[['Date','PROVINSI','Daily_Case','geometry']]
daily_case = daily_case.query('Date == "01/05/2021"')
daily_case.head()
```

<h1>PLOTTING PETA

```python
daily_case.plot(figsize=(20,10))
```

```python
import mapclassify as mc

scheme = mc.FisherJenks(daily_case['Daily_Case'], k=5)

ax = gplt.choropleth(
    daily_case, hue='Daily_Case', figsize=(20,10),
    edgecolor='black', linewidth=1, scheme=scheme,
    cmap='Greens', legend=True, legend_kwargs={'loc': 'lower left'}
)

ax.set_title(
    "Peta Persebaran Kasus Positif Covid-19 per Tanggal 01 Mei 2021",
    fontsize=18
)
```
